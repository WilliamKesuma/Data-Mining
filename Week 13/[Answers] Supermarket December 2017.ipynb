# MARKET BASKET ANALYSIS

```python
# Create a pivot table representing transactions, with each row as a transaction and columns as products,
# filling in zeros where a product is not present in a transaction
basket = df.groupby(['NO_BKT', 'KEL'])['KEL'].count().unstack().reset_index().fillna(0).set_index('NO_BKT')
basket.head()
```

```python
# Step 1: Create a pivot table to represent transactions, filling in zeros for missing values
basket = df.pivot_table(index='NO_BKT', columns='KEL', aggfunc='size', fill_value=0)

# Step 2: Transform the matrix to binary by converting values greater than 1 to 1
basket[basket > 1] = 1

# Step 3: Display the first few rows of the transformed pivot table
basket.head()
```

```python
# Step 1: Apply the Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(basket, min_support=0.02, use_colnames=True)

# Step 2: Display the frequent itemsets along with their support values
print("Frequent Itemsets:")
print(frequent_itemsets.head(35))
```

Based on the analysis of support values for various products, the following business recommendations can be made:

1. **Prioritize high-support products**: Items like Pampers, MINUMAN (beverages), and SABUN.M (soap) are frequently purchased together and should be the main focus of marketing efforts. Offering promotions, discounts, or loyalty programs centered around these products could enhance customer engagement.

2. **Leverage cross-selling opportunities**: Products with moderate to high support, such as DETERJN (detergent), KOPI (coffee), and SABUN.C (soap), are often bought together. This suggests a chance to create bundles or package deals that encourage customers to purchase related items. For example, offering discounts on coffee and snacks or detergent and fabric softener could drive sales.

3. **Seasonal promotions**: Products like PAMPERS, SABUN.M, and MIE (instant noodles) with high support could be included in seasonal promotions or flash sales, as they are frequently purchased and have broad customer appeal. Aligning these offers with holiday seasons or special events can further boost sales.

4. **Targeted marketing for low-support items**: Products such as ROTI.B (bread) or ROKOK (cigarettes) with low support should be marketed to specific customer segments. Tailoring marketing campaigns for these niche products—such as targeting ads towards families for bread or regular smokers for cigarettes—can help attract their intended audience.

5. **Optimize inventory management**: High-support products like Pampers and MINUMAN should be prioritized in inventory management to ensure they are always in stock and prominently displayed. Low-support items may be managed using just-in-time inventory strategies or through special promotions to move stock.

6. **Enhance customer loyalty**: For products with both high support and loyal customer bases, such as SABUN.M, consider creating loyalty programs. Customers could earn rewards for purchasing these items regularly, which increases the likelihood of repeat purchases.

By understanding how frequently products are purchased together and the relationships between them, these strategies can enhance customer satisfaction, boost sales, and improve inventory and marketing management.
